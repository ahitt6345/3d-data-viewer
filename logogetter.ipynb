```python
import os

# Create a directory to save the logos
if not os.path.exists('logos'):
    os.makedirs('logos')
```

```python
# List of companies
companies = []

import csv

csv_file_path = 'Backend/De No Finance 2.csv'

# Read the CSV file and extract company names
with open(csv_file_path, mode='r', newline='', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)  # Skip the header row
    for row in csv_reader:
        company_name = row[0]
        company_website = row[2]
        companies.append((company_name, company_website))

# Print the extracted company names (for verification)
for company in companies:
    print(company)

# Optionally, you can also use pandas to read the CSV file and extract the company names
import pandas as pd

# Read the CSV file
df = pd.read_csv(csv_file_path)

# Extract company names
companies_pandas = df['Companies'].tolist()

# Print the extracted company names (for verification)
for company in companies_pandas:
    print(company)
```

```python
import requests
# Function to get logo URL from Clearbit API
def get_logo_url(domain):
    api_url = f"https://logo.clearbit.com/{domain}"
    return api_url

# Function to download and save the logo
def download_logo(url, company_name):
    response = requests.get(url)
    if response.status_code == 200:
        file_path = os.path.join("logos", f"{company_name}.png")
        with open(file_path, "wb") as file:
            file.write(response.content)
        print(f"Downloaded logo for {company_name}")
    else:
        print(f"Could not find logo for {company_name}")

# Main loop to process each company
for company_name, company_url in companies:
    # if logo already exists, skip
    if os.path.exists(f"logos/{company_name}.png"):
        print(f"Logo for {company_name} already exists. Skipping...")
        continue
    logo_url = get_logo_url(company_url)
    response = requests.get(logo_url)
    if response.status_code == 200:
        download_logo(logo_url, company_name)
    else:
        print(f"No logo found for {company_name}")

print("Finished downloading logos.")
```

```python
#print a list of the companies we could not find logos for
companies_without_logos = [company_name for company_name, _ in companies if not os.path.exists(f"logos/{company_name}.png")]
print("Companies without logos:")
for company in companies_without_logos:
    print(company)
```
